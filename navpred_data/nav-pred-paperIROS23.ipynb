### Base Data set

#### Base Data set stats

```python
base_data_set_stats = base_data_set.describe()
base_data_set_stats['average_time_diff'] = base_data_set_stats['average_time_diff'].astype('int64') 
base_data_set_stats.round(2)
```

#### Base Data set robot composition

```python
data = base_data_set

burger = data[data["robot_max_speed"]==0.22]
ridgeback = data[data["robot_max_speed"]==1.10]
jackal = data[data["robot_max_speed"]==2.00]

print("Burger:",len(burger),"entries /",round(((len(burger)/len(data))*100),2),"%")
print("Ridgeback:",len(burger),"entries /",round(((len(ridgeback)/len(data))*100),2),"%")
print("Jackal:",len(jackal),"entries /",round(((len(jackal)/len(data))*100),2),"%")
```

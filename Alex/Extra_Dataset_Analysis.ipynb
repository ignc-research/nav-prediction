#### Extra Data set stats

```python
extra_data_set_stats = extra_data_set.describe()
extra_data_set_stats['average_time_diff'] = extra_data_set_stats['average_time_diff'].astype('int64')
extra_data_set_stats.round(2)
```

Extra Data set planner composition



#### Extra Data set robot composition

```python
data = extra_data_set

burger = data[data["robot_max_speed"]==0.22]
ridgeback = data[data["robot_max_speed"]==1.10]
jackal = data[data["robot_max_speed"]==2.00]

print("Burger:",len(burger),"entries /",round(((len(burger)/len(data))*100),2),"%")
print("Ridgeback:",len(ridgeback),"entries /",round(((len(ridgeback)/len(data))*100),2),"%")
print("Jackal:",len(jackal),"entries /",round(((len(jackal)/len(data))*100),2),"%")
```
